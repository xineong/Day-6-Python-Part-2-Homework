```
"""
Created on Sun Sep 27 14:11:36 2020

@author: Xine Ong
"""

pastuserscores = []

####################################################################################################################################################################

def run_quiz():
  
  print("\nWELCOME TO THE QUIZ !!!\nJust input the letter of your answer for each question. Let's begin!\n")
  
  quizfile = open("quiz.txt","r")
  filelist = quizfile.readlines()
  quizfile.close()    

  score = 0
  
  ## building our list of questions in tuple form
  
  quizlist = []
  for k, line in enumerate(filelist):
    tuple_complete = False
    line1 = line.strip()

    ## build a tuple for each question

    if k % 6 == 0:
        question = line1
    if k % 6 == 1:
        choice1 = line1
    if k % 6 == 2:
        choice2 = line1
    if k % 6 == 3:
        choice3 = line1
    if k % 6 == 4:
        choice4 = line1
    if k % 6 == 5:
        answer_code = line1
        tuple_complete = True
    if tuple_complete:
        tuple1 = (question, choice1, choice2, choice3, choice4, answer_code)
        quizlist.append(tuple1)

  ## quiz proper

  for index, q in enumerate(quizlist):
    print(f"{index+1}. {q[0]} \n\t{q[1]} \n\t{q[2]} \n\t{q[3]} \n\t{q[4]}")
    useranswer = input("Answer: ")
    if useranswer == q[5]:
      print ("That's correct!\n")
      score += 1
    else:
      print (f"Sorry, that is incorrect. The correct answer is [{q[5]}].\n")

  ## user score at the end of the quiz

  quizlen = len(quizlist)
  rate = float(score / quizlen) * 100

  if int(rate) >= 70:
    pass_fail = "Congrats, you passed!"
  else:
    pass_fail = "Sorry, you failed."

  print(f"\n\nTOTAL SCORE: {str(score)} out of {quizlen}")
  print(f"You got {int(rate)}%. {pass_fail}")


  ## displays past 10 user-scores

  pastuserscores.insert(0,(user, int(rate)))

  if len(pastuserscores) == 11:
    pastuserscores.pop()

  print("\nPast 10 quiz takers and scores:")
  for userscore in pastuserscores:
    print (f"Username: {userscore[0]} \t Score: {userscore[1]}%")


  ## asks whether another player wants to run the game again

  global run_option
  quizagain = 0
  while quizagain == 0:
   
    quizagain = input ("\nWould another player like to try? \n[Y] Enter Y for YES \t[N] Enter N for NO \n")

    if quizagain.lower() == "y":
      quizagain = 1
    elif quizagain.lower() == "n":
      print("Okay, thank you for playing!")
      run_option = "q"
      quizagain = 1
    else:
      print("Sorry, I don't understand. Please try again.")
      quizagain = 0
          

#########################################################################################################################################################################         

def modify_quiz():

  global run_option
  modify = 0

  while modify == 0:
  
    modify = input("\nWould you like to add or remove questions from the quiz? \n[1] Enter 1 to add a question \n[2] Enter 2 to remove a question \n[3] Enter 3 to go back to main menu \n")

    ## add new question

    if modify == "1":
      question = input("\nPlease input question to be added to the quiz:\n") + "\n"
      choice1 = "a. " + input("For the multiple choices: \nPlease input choice (a.): ") + "\n"
      choice2 = "b. " + input("Please input choice (b.): ") + "\n"
      choice3 = "c. " + input("Please input choice (c.): ") + "\n"
      choice4 = "d. " + input("Please input choice (d.): ") + "\n"
      answer_code = input("Please enter the letter of the correct answer: ") + "\n"

      quizfile = open("quiz.txt","a")
      quizfile.write(question)
      quizfile.write(choice1)
      quizfile.write(choice2)
      quizfile.write(choice3)
      quizfile.write(choice4)
      quizfile.write(answer_code)
      quizfile.close

      print ("New question added.")
      modify = 0

    ## remove question

    elif modify == "2":

      quizfile = open("quiz.txt","r")
      filelist = quizfile.readlines()
      quizfile.close()   

      print("\nThis is the list of questions currently on the quiz: ")
      for index, q in enumerate(filelist):
        if index % 6 == 0:
          print(f"[{index//6}] {q.strip()}")
      questionremove = input("\nWhich question would you like to remove? \nPlease enter the question number of your choice: ")
      qr = int(questionremove) * 6
      qr_range = [qr, qr+1, qr+2, qr+3, qr+4, qr+5]

      quizfile = open("quiz.txt","w")
      for index, q in enumerate(filelist):
        if index not in qr_range:
          quizfile.write(q)
      quizfile.close()    
      
      print ("Question removed.")
      modify = 0

    ## go back to main menu
    elif modify == "3":
      modify = 1
  
    ## catch input not in choices
    else:
      print ("Sorry, I don't understand. Please try again.")
      modify = 0


#########################################################################################################################################################################

# START OF PROGRAM

run_option = " "

while run_option != "q":
  user = input("\nHello! What's your name? ")
  run_option = input(f"\nHi {user.title()}! What do you want to do today?\n[1] Enter 1 if you want to take the quiz\n[2] Enter 2 if you want to modify the quiz\n[Q] Enter q if you want to quit the game\n")

  if run_option == "1":
    run_quiz()
       
  elif run_option == "2":
    modify_quiz()
  
  elif run_option.lower() == "q":
    print("\nThank you for playing. Goodbye!")

  else:
    print("\nYour choice is not found in the options available. Please try again.")

# END OF PROGRAM
```
